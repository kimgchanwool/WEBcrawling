```python
import requests
from bs4 import BeautifulSoup
```

```python
def spider(url, used_list = []):
    try:
        soup = BeautifulSoup(requests.get(url, timeout=1).text)
        Atags = soup.find_all('a')
        while Atags: #que형식이다. 사용한 것 버리기.
            now_link = Atags.pop(0)['href']
            if now_link != '#':
                if 'http' in now_link and '.exe'  not in now_link and '.dmg'  not in now_link:
                    now_link
                    if now_link.split('://')[-1] not in used_list:
                        print(now_link)
                        used_list.append(now_link.split('://')[-1])
                        spider(now_link)
    except requests.Timeout:
    # back off and retry
        pass
 
    except requests.ConnectionError:
        pass
    except:
        pass
```

```python
spider('https://www.naver.com')
```
